```python
import numpy as np
```

```python
import time
```

```python
# 1. Array Creation
array_1d = np.array([10, 20, 30, 40, 50])
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
```

```python

print("1D Array:", array_1d)
print("2D Array:\n", array_2d)
```

```python
# 2. Indexing and Slicing
print("First element:", array_1d[0])
print("Slice (index 1 to 3):", array_1d[1:4])
print("2D element:", array_2d[1, 2])
```

```python
# 3. Mathematical Operations
print("Addition:", array_1d + 5)
print("Multiplication:", array_1d * 2)
```

```python
# 4. Statistical Operations
print("Mean:", np.mean(array_1d))
print("Median:", np.median(array_1d))
print("Standard Deviation:", np.std(array_1d))
print("Sum:", np.sum(array_1d))
```

```python

# 5. Axis-wise Operations
print("Column Sum:", np.sum(array_2d, axis=0))
print("Row Sum:", np.sum(array_2d, axis=1))
```

```python
# 6. Reshaping
reshaped = array_1d.reshape(5, 1)
print("Reshaped Array:\n", reshaped)
```

```python

# 7. Broadcasting
broadcast_result = array_2d + np.array([10, 20, 30])
print("Broadcasting Result:\n", broadcast_result)
```

```python

# 8. Save and Load NumPy Array
np.save("data_saved.npy", array_1d)
loaded_array = np.load("data_saved.npy")
print("Loaded Array:", loaded_array)
```

```python
# 9. Performance Comparison
size = 1000000
```

```python
python_list = list(range(size))
numpy_array = np.arange(size)
```

```python
start = time.time()
python_sum = sum(python_list)
end = time.time()
print("Python List Sum Time:", end - start)
```

```python
start = time.time()
numpy_sum = np.sum(numpy_array)
end = time.time()
print("NumPy Array Sum Time:", end - start)
```
